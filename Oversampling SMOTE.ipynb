```python
import numpy as np
import pandas as pd
from collections import Counter
from sklearn.datasets import make_classification
from imblearn.over_sampling import SMOTE 
```

```python
df = pd.read_excel(
    r'C:\Users\admin\Jupyter project\1 Project NN\3 Данные из финансового модулятора\output.xlsx'
)
```

```python
df = df[df['kind'] == 'ИО']
```

```python
factbad = 90

df['factbad'] = np.where(df['Макс просрочка'] > factbad, 1, 0)

# df.loc[df['Макс просрочка'] > factbad, 'factbad'] = 1
# df.loc[df['Макс просрочка'] <= factbad, 'factbad'] = 0
```

```python
statusSearch = [c for c in df.columns if c.lower() == 'factbad']
status = df[statusSearch]
```

```python
predSearch = [c for c in df.columns if c.lower()[:2] == 'cf']
```

```python
predSearch = [
    'cf (total_equity - intangibleAssets) / total_liabilities',
    'cf (total_equity - total_noncurrent_assets) / total_current_assets',
    'cf 6000 / Assets',
    'cf 6000 on cash and stocks',
    'cf ATR',
    'cf EBIT on current_liabilities',
    'cf EBIT on total_assets',
    'cf EBIT/interest log10',
    'cf OPL / 6000',
    'cf OPL / Equity',
    'cf OPL on liabilities',
    'cf OPL on total_assets',
    'cf OPL on total_equity',
    'cf PBT on COGS',
    'cf PBT on total_assets',
    'cf PBT on total_equity',
    'cf ROA',
    'cf accounts receivable on total_liabilities',
    'cf cash and stocks / assets',
    'cf currentRatio',
    'cf currentRatioAltman',
    'cf netProfit on expenses',
    'cf netProfit on liabilities',
    'cf netProfit on total_equity',
    'cf total_current_assets on total_liabilities',
    'cf total_current_liabilities on total_assets',
    'cf total_current_liabilities rate',
    'cf total_equity on total_liabilities',
    'cf total_liabilities on total_assets',
    'cf total_liabilities on total_equity_and_liabilities',
    'cf total_noncurrent_liabilities rate',
    'cf workingCapital on liabilities',
    'cf workingCapital on total_assets'
]
```

```python
pred = df[predSearch]
```

```python
pred.describe()
```

```python
pred[['cf currentRatio', 'cf currentRatioAltman']].describe()
```

```python
listCf = [c for c in pred.columns if c.lower()[:2] == 'cf']
newdf = pd.DataFrame()
for i in listCf:
    newdf[i] = pred[i]
    
for cat in newdf:
    quantileCat = newdf[cat].quantile(.75)
    quantileCatNegative = quantileCat*-1
    newdf.loc[newdf[cat] > quantileCat, cat] = quantileCat
    newdf.loc[newdf[cat] < quantileCatNegative, cat] = quantileCatNegative
```

```python
newdf.describe()
```

```python
y_new = status.to_numpy()
y_new = y_new[:,0]
X_new = newdf.to_numpy()
sm2 = SMOTE(sampling_strategy=0.5, random_state=42, k_neighbors=4)
X_newres, y_newres = sm2.fit_resample(X_new, y_new)
print('Original dataset shape %s' % Counter(y_new))
print('Resampled dataset shape %s' % Counter(y_newres))
```

```python
part1 = pd.DataFrame(X_newres)
part2 = pd.DataFrame(y_newres)
concat_part = pd.concat([part2, part1], axis=1, join="inner")
```
